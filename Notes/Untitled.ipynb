```python
import string

ma         = text.as_matrix()[:,0]
dictionary = set()
freq       = dict()

for line in ma:
    words = [ x.lower() for x in line.split() ] 
    for word in words:
        word = word.translate(string.maketrans("",""), string.punctuation)
        if word in dictionary:
            freq[word] += 1
        else:
            freq[word] = 1
            dictionary.add(word)
```

```python
freq
```

```python
threshold = 5
for key, value in sorted(freq.iteritems(), key=lambda (k,v): (v,k)):
    if value >= threshold: 
        print "%s: %s" % (key, value)
```
